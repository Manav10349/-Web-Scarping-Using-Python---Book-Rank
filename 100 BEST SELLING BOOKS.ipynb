```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
table=soup.find('div',id='article-body-blocks')
```

```python
table=table.find('tbody')
```

```python
rows=table.find_all('tr',style='')
rows
```

```python
column_info=[]
for i in rows:
    cols=i.find_all('td')
    book_info=[c.text for c in cols]
    column_info.append(book_info)
print(column_info)
```

```python
Rank=[]
Title=[]
Author=[]
Volume_Sales=[]
Publisher=[]
Genre=[]
for i in column_info:
    Rank.append(i[0])
    Title.append(i[1])
    Author.append(i[2])
    Volume_Sales.append(i[3])
    Publisher.append(i[4])
    Genre.append(i[5])
    
```

```python
Data={'Rank':Rank,'Title':Title,'Author':Author,'Volume Sales':Volume_Sales,'Publisher':Publisher,'Genre':Genre}
```

```python
book_info=pd.DataFrame(Data)
book_info
```

```python
book_info.to_csv('books.csv')
```
